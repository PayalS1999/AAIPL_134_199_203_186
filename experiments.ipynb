Check if the few shot examples generated by Gemini stick to token length limits

```python
import json

with open("few_shot.json") as f:
    examples = json.load(f)
```

```python
len(examples)
```

```python
correct_dict = {}
for example in examples:
    if example["topic"] not in correct_dict:
        correct_dict[example["topic"]] = []
    correct_dict[example["topic"]].append({k: v for k, v in example.items() if k not in ["topic"]})

print(json.dumps(correct_dict, indent=2))
```

```python
with open("correct_few_shot.json", "w") as f:
    json.dump(correct_dict, f)
```
